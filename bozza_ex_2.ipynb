```python
lis_df = []
for i in list_df:
    lis_df.append(list(map(int,(i[0].split())[1:])))
    
```

```python
new_lis_df =[]
for i in lis_df:
    i = i[::-1]
    new_lis_df.append(i)
```

```python
parent = dict()
rank = dict()
for i in range(1,len(lis_df)):
    parent[i] = i
    rank[i] = 0

def find(vertice):
    if parent[vertice] != vertice:
        parent[vertice] = find(parent[vertice])
    return parent[vertice]

def union(vertice1, vertice2):
    root1 = find(vertice1)
    root2 = find(vertice2)
    if root1 != root2:
        if rank[root1] > rank[root2]:
            parent[root2] = root1
    else:
        parent[root1] = root2
    if rank[root1] == rank[root2]: rank[root2] += 1
```
